#### Everything we need to know to understand our datasets :

Race distances are measured in kilometers and not laps. Most Formula 1 races take between 50-70 laps to complete the required 305 kilometers. Circuits that are longer will require fewer laps to reach the required distance of 305 kilometers, and shorter circuits will take more laps to reach the requirement.

```python
def split_it(x) :
    list_hm = x.split(':')
    if( x != 'nan') :
        return round(float(list_hm[0]) + (float(list_hm[1])/60),6)
    else :
        return pd.NA
```

```python
results['fastestLapTime'] = results['fastestLapTime'].astype(str)
```

```python
(results['fastestLapTime'].map(type) != str).sum()
```

```python
results['fastestLapTime'] = results['fastestLapTime'].apply(split_it)
```

```python
results.fastestLapTime
```

```python
results.time_total
```

The column that contains the best speed from all the laps is in upm (unit miles per hour), we will convert them in km/h because m/s is the international unit of speed (SI unit of speed)

```python
results['fastestLapSpeed'] = round(results['fastestLapSpeed']*1.609344,6)
```

```python
# we will rename the column 
rename_speed = {'fastestLapSpeed' : 'fastestLapSpeed_km/h'}

results.rename(columns = rename_speed, inplace = True)
```

```python
results['num_fastestLap'] = results['num_fastestLap'].astype('Int64')
results['num_car'] = results['num_car'].astype('Int64')
results['start_pos'] = results['start_pos'].astype('Int64')
results['final_pos'] = results['final_pos'].astype('Int64')
```

```python
to_drop = ['time_ms']

results.drop(columns=to_drop,axis = 1,inplace = True)
```
