```python
df_constructors.join(df_drivers.set_index("nationality"), on = "nationality", lsuffix= True)
```

```python
nb_circuit_pays = df_circuit.groupby(by = "country").count().sort_values(by = ["circuitId", "country"], ascending = [False, True]).reset_index()[["country", "circuitId"]]
nb_circuit_pays.head(5)
```

```python
nb_constructeur_pays = df_constructors.groupby(by = "nationality").count().sort_values(by = ["constructorId", "nationality"], ascending = [False, True]).reset_index()[["nationality", "constructorId"]]
nb_constructeur_pays.head(5)
```

```python
nb_pilote_pays = df_drivers.groupby(by = "nationality").count().sort_values(by = ["driverId", "nationality"], ascending = [False, True]).reset_index()[["nationality", "driverId"]]
nb_pilote_pays.head(5)
```

```python
nb_constructeur_pilote_pays = nb_constructeur_pays.set_index("nationality").join(nb_pilote_pays.set_index("nationality"), on = "nationality").reset_index()
nb_constructeur_pilote_pays
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax2 = ax.twinx()

x = nb_constructeur_pilote_pays["nationality"]
y = nb_constructeur_pilote_pays["constructorId"]
z = nb_constructeur_pilote_pays["driverId"]
  
X_axis = np.arange(len(x))
  
ax.bar(X_axis - 0.2, y, 0.4, color = "red", label = 'nombre de constructeurs')
ax2.bar(X_axis + 0.2, z, 0.4, color = "blue", label = 'nombre de pilotes')

ax.set_xticklabels(x, rotation = 90)
plt.xticks(X_axis, x, rotation = "vertical")
plt.xlabel("nationality")

fig.legend()
plt.show()
```
